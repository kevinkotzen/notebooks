```python
%load_ext autoreload
%autoreload 2
%matplotlib widget
import heartpy as hp
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal, interpolate
import copy
import os 
import pandas as pd
os.chdir('/home/kkotzen/research/PPG_sleepstaging_orion3/')
from pathlib import Path
from src.parsing.SHHSParser import SHHSParser
from scipy.signal import butter, sosfiltfilt, resample, cheby1
from src.parsing.utils.extract_numpy_from_edf import extract_numpy_from_edf
from joblib import Parallel, delayed
import tqdm

fs_ecg = 125
fs_resp = 10
fs_target = 125
destination = "/home/kkotzen/databases/shhs/signals/npy/ECG+ABDO RES/"


def add_signals(patient):
    dl = SHHSParser()
    ecg = extract_numpy_from_edf(dl.path_edf(patient), 'ECG')

    resp = extract_numpy_from_edf(dl.path_edf(patient), 'ABDO RES')
    sos_resp = cheby1(8, 0.001, 1 / (fs_resp / 2), 'lowpass', output='sos')
    resp_filt = sosfiltfilt(sos_resp, resp, axis=0)
    resp_resample = resample(resp_filt, int(len(resp_filt)/(fs_resp/fs_target)))
    
    addition =  ecg/np.mean(np.abs(ecg)) + resp_resample/np.mean(np.abs(resp_resample))
    addition = (addition / np.mean(np.abs(addition)))/10
    return addition

def process(patient):
    try:
        np.save(f"{destination}shhs1-{patient}-ECG+ABDO RES.npy", add_signals(patient))
    except:
        print(f"Failed {patient}")
        
dl = SHHSParser()
Parallel(n_jobs=200)(delayed(process)(patient) for patient in tqdm.tqdm(dl.database_all_patient_IDs))
# process(dl.database_all_patient_IDs[888])
```
