```python
from scipy.stats import ks_2samp
import pandas as pd
import numpy as np

def load_results(hash_code):
    results = pd.read_excel(f"/home/kkotzen/tmp/results/{hash_code}.xlsx")
    return results['Kappa:'].values

def calculate_metrics(name, hash_no_pretrain, hash_with_pretrain):
    no_pretrain = load_results(hash_no_pretrain)
    with_pretrain = load_results(hash_with_pretrain)
    sig = ks_2samp(no_pretrain, with_pretrain).pvalue
    print(name, " : ", np.median(no_pretrain), ' & ',  np.median(with_pretrain), " & ", np.median(with_pretrain) - np.median(no_pretrain), " & ", sig)

print("Without Pretrain vs with Pretrain")
calculate_metrics('BM-FE', "f15ed8d33f944b0caa6c7f1f6fe437a9", "222108c64d72451287876b465e20e9bb")
calculate_metrics('BM-DTS', "2cf0c668f85f4c6aa8c16dd9fb4912f2", "f296a562b5fa472fbee35da5d49648a8")
calculate_metrics('SleepPPG-Net', "1e42d89a3eb84612a1e0b9119e4f2149", "66c27d8a81ee4a1e87ee3a4f8b0a14b8")

print("PPG vs ECG")
calculate_metrics('BM-FE', "8753c4563d914db1ac9319a92ff0edc8", "222108c64d72451287876b465e20e9bb")
calculate_metrics('BM-DTS', "7f2ab8658ab14427a24e5bf93e49efe7", "f296a562b5fa472fbee35da5d49648a8")
calculate_metrics('SleepPPG-Net', "626828885c7040678b0a7a8affaac635", "66c27d8a81ee4a1e87ee3a4f8b0a14b8")

```
